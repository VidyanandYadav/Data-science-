#### Important DataFrame Functions

```python
# astype
ipl.info()
```

```python
ipl["ID"].astype("int32")
# This will changes the ID dtype from int64 to int32 which reduces the memory in system 
```

```python
# To make permanent changes that ID ko int64 to int32 do this
ipl["ID"] = ipl["ID"].astype("int32")
ipl.info()
# See below now ID's dtype has been changed permanently and also memory usage reduces from 148.6 to 144.9
```

###### One more example is that season in ipl is intteger such as 2008,2009,2010.. 2022 which is like same category so conerting its dtype into category reduces the memory usage

```python
ipl["Season"] = ipl["Season"].astype("category")
ipl.info()
# See in output that memory usage has been reduced to 139.0 from 144.9
```

```python
# You can also do the the same for Team1 Team2 Toss Winner etc
ipl["Team1"] = ipl["Team1"].astype("category")
ipl["Team2"] = ipl["Team2"].astype("category")
ipl.info()
# See below the dtype of Team1 and Team2 is changed to category and also memory reduced to 127
```
