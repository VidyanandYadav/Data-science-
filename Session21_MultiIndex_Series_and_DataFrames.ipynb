#### Series is 1D and DataFrames are 2D objects
##### But why?
##### And what exactly is index?

```python
# can we have multiple index? Let's try
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
a = pd.Series([1,2,3,4,5,6,7,8],index=index_val)
a
```

```python
# The problem?
a['cse']
```

```python
# The solution -> multiindex series(also known as Hierarchical Indexing)
# multiple index levels within a single index
```

```python
# how to create multiindex object
# 1. pd.MultiIndex.from_tuples()
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
multiindex = pd.MultiIndex.from_tuples(index_val)
multiindex.levels[1]
# 2. pd.MultiIndex.from_product()
pd.MultiIndex.from_product([['cse','ece'],[2019,2020,2021,2022]])
```

```python
# level inside multiindex object
```

```python
# creating a series with multiindex object
s = pd.Series([1,2,3,4,5,6,7,8],index=multiindex)
s
```

```python
# how to fetch items from such a series
s['cse']
```

```python
s['cse',2021]
```

#### Stacking and Unstacking

```python
# In unstack for there is multiIndex row(so inner row as it is but outer row becomes columns just below original columns)
branch_df1
```

###### NOTE : This two indexing cse,ece and 2019 to 2022 is called as Level 1 and Level 2 indexing

```python
branch_df1.unstack()
```

```python
# What happens when we use one more unstack() -->> The remaining two rows ece and cse will distribute to each year
branch_df1.unstack().unstack()
# Also this is no more any dataFrame becausethere is no any row ans=d hence it is now changes into series 
```

Now when you use stack then inner column changes into row and also note that stack doesn't applicable on series

```python
branch_df1.unstack().stack()
```

```python
# Now if we again do stack then remaining column(Level 0) wala will converted into row for each years ..>> see beloe
branch_df1.unstack().stack().stack()
```

#### Long Vs Wide Data

Wide format is where we have a single row for every data point with multiple columns to hold the values of various attributes.

Long format is where, for each data point we have as many rows as the number of attributes and each row contains the value of a particular attribute for a given data point.

```python
# melt -->> Wide data format ko long data format me convert karta hai
# pivot-->> Long data format ko wide data format me convert karta hai 
```

```python
# melt -->> Simple example branch
pd.DataFrame({'cse':[120]})
```

```python
pd.DataFrame({'cse':[120]}).melt()
```

```python
# melt -->. branch with year
pd.DataFrame({'cse':[120],'ece':[100],'mech':[50]})
```

```python
pd.DataFrame({'cse':[120],'ece':[100],'mech':[50]}).melt()
```

```python
pd.DataFrame({'cse':[120],'ece':[100],'mech':[50]}).melt(var_name = 'Branch',value_name = 'No of Students')
```

```python
pd.DataFrame(
    {
        'branch':['cse','ece','mech'],
        '2020':[100,150,60],
        '2021':[120,130,80],
        '2022':[150,140,70]
    }
)
```

```python
pd.DataFrame(
    {
        'branch':['cse','ece','mech'],
        '2020':[100,150,60],
        '2021':[120,130,80],
        '2022':[150,140,70]
    }
).melt()
# This doesn't gives any sense coz hame branch ko row nahi banana tha but 2020,2021,2022 ko krna tha so use id_vars = []
# And inside [] pass that parameter for which you don,t want to change it into rows
```

```python
pd.DataFrame(
    {
        'branch':['cse','ece','mech'],
        '2020':[100,150,60],
        '2021':[120,130,80],
        '2022':[150,140,70]
    }
).melt(id_vars=['branch'],var_name = 'Year',value_name = 'Students')
```

Question: Let you have to make a dataset using these two datasets death and confirm in which there are four columns namely 1.Country 2.Date 3.Confirm cases 4.Death cases and this dataset is absolutely in long data format so you've to use melt function

```python
death = death.melt(id_vars=['Province/State','Country/Region','Lat','Long'],var_name = 'date',value_name = 'num_deaths')
# id_vars=['Province/State','Country/Region','Lat','Long'] is used to keep these var as in columns don't convert in rows
confirm = confirm.melt(id_vars=['Province/State','Country/Region','Lat','Long'],var_name = 'date',value_name = 'num_cases')
```
