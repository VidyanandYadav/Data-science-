Merging

NOTE: By default there is inner joining

###### Joining means -->> Let we've to join two datasets or tables regs and students in which StdID is common in both(common column is necessay for joining ).Here regs (2 colmns StdID and CollegeID) and students(2 clmns StdID and Name) so joined datsets have columns such as StdID Name CollegeID.
###### There are four types of joining  a. left b.right c.outer  d.inner

```python
# (a).inner join -->> You have to take or select that column which columns in both datasets or tables
# NOTE: In inner joining results me wahi chize aayengi jo dono datasets or table me present ho
students.merge(regs,how = 'inner',on = 'student_id')
```

###### (b).left join -->> Output me wo sari chize aayengi to common ho aur left wale mein ho but not that wala jo right me ho but left me nhi ho.

```python
# Here we join courses and regs on the basis of course_id bcoz it is common in both
courses.merge(regs,how = 'left', on = 'course_id')
# See on output Numpy and C++ are also printed as it is present in left but not in right
```

```python
# Here we just making datasets more comfortable so that analysis can be done easily -->> it's not a part of any question
temp_df = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Ankit','Rahul'],
    'partner':[28,26,17]
})
students = pd.concat([students,temp_df],ignore_index = True)
```

###### (c).right join -->>Output me wo sari chize aayengi to common ho aur right wale mein ho but not that wala jo left me ho but right me nhi ho.

```python
# Here we join students and regs on the basis of student_id bcoz it is common in both
students.merge(regs,how = 'right', on = 'student_id')
```

###### (d).Outer Join (Full Join) -->> Output me wo sari chize aayengi jo common ho aur right wale mein ho aur left wale me bhi ho.

```python
# outer join
students.merge(regs,how = 'outer' , on = 'student_id').tail(10)
# tail(10) is used coz last me hi diiference dikh rha hai 
```

```python
# 1.Find total revenue generated
total = regs.merge(courses,how = 'inner',on = 'course_id')['price'].sum()
total
# When using['price'] by this from all columns of datasets only price column will be shown
```

```python
# 2.Find month by month revenue
temp_df = pd.concat([nov,dec],keys = ['November','December']).reset_index()
temp_df.merge(courses,on='course_id').groupby('level_0')['price'].sum()
```

```python
# 3.Print the registration table
# columns -->> names -->> course -->> price (HINT: Here you have to merge three tables)
# NOTE : In this question requires 3 clmns are present in 3 diff datasets so u have to merge two times in correct manner
regs.merge(students,on = 'student_id').merge(courses,on = 'course_id')[['name','course_name','price']]
```

```python
# 4.Plot bar chart for revenue/course
regs.merge(courses,on = 'course_id').groupby('course_name')['price'].sum().plot(kind = 'bar')
```

```python
regs.merge(courses,on = 'course_id').groupby('course_name')['price'].sum().plot(kind = 'pie')
```

```python
# 5.Find students who enrolled in both the months
common_student_id = np.intersect1d(nov['student_id'],dec['student_id'])
common_student_id
```

```python
students[students['student_id'].isin(common_student_id)]
```

```python
# 6.Find course that got no enrollment -->> means that course_id which are in courses but not in registrations
# courses['course_id']
# regs['course_id']
course_id_list = np.setdiff1d(courses['course_id'],regs['course_id'])
courses[courses['course_id'].isin(course_id_list)]
```

```python
# 7.Find students who did not enroll into any courses -->> which are in students table but not in registration table
student_id_list = np.setdiff1d(students['student_id'],regs['student_id'])
students[students['student_id'].isin(student_id_list)]
```

NOTE : setdiff1d means it gives set difference operation i.e A-B means element which are in A but not present in B

```python
# 8.Print students name -->> partner name for all enrolled students
# Here we use fifth joining called self join
students.merge(students,how = 'inner',left_on = 'partner',right_on = 'student_id')[['name_x','name_y']]
```

```python
# 9.Find top 3 students who did most number enrollments
regs.merge(students,on = 'student_id').groupby(['student_id','name'])['name'].count().sort_values(ascending = False).head(3)
```

```python
# 10.Find top 3 students who spent most amount of money on courses
regs.merge(students,on = 'student_id').merge(courses,on ='course_id').groupby(['student_id','name'])['price'].sum().sort_values(ascending = False).head(3)
```

```python
# IPL Problems
# Find top 3 stadiums with highest sixes/match ratio
# Find Orange cap holder of all seasons
```

```python
temp_df = delivery.merge(matches,left_on ='match_id',right_on = 'id')
```

```python
six_df = temp_df[temp_df['batsman_runs'] == 6]
```

```python
# Stadium sixes
num_sixes = six_df.groupby('venue')['venue'].count()
```

```python
num_matches = matches['venue'].value_counts()
```

```python
(num_sixes/num_matches).sort_values(ascending = False).head(3)
```

```python
# Orange Cap holder for each seasons
```

```python
# In delivery we've all data to calculate batsman runs but there is no any data for seasons so for seasons we want matches data
# So to get both data we have to merge delivery and matches which was done in above cell and stored in temp_df
temp_df.groupby(['season','batsman'])['batsman_runs'].sum().reset_index().sort_values('batsman_runs',ascending = False).drop_duplicates(subset = ['season'],keep='first').sort_values('season')
```
