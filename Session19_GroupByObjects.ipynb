#### Aggregate Methods

genres.sum() gives sum of all numeriacl data which in sense not too beneficial such as sum of Runtime of all action movies has not any sense but if we do avg it's good and same for IMDB rating but in case of no of votes sum is helpful. So for diff parameter we have diff need of operation that's why we use agg method.

```python
# passing dictionary
genres.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# passing list -->> for each numerical column we want more than one operation so pass a list instead of dict
# genres.agg(['min','max','mean']) -->> having some issues
```

```python
# Adding both the Syntax
genres.agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# looping on groups
df = pd.DataFrame(columns = movies.columns)
for group,data in genres:
    top_rated = data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
    df = pd.concat([df,top_rated])
df
# NOTE: Here are some changes coz df = df.append(data[data["IMDB_Rating"]==data["IMDB_Rating"].max()]) has been removed 
# In the new version of Jupyter Notebook
```
