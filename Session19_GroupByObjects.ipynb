#### Aggregate Methods

genres.sum() gives sum of all numeriacl data which in sense not too beneficial such as sum of Runtime of all action movies has not any sense but if we do avg it's good and same for IMDB rating but in case of no of votes sum is helpful. So for diff parameter we have diff need of operation that's why we use agg method.

```python
# passing dictionary
genres.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# passing list -->> for each numerical column we want more than one operation so pass a list instead of dict
genress = movies.groupby(['Director', 'Star1'])[movies.select_dtypes(include='number').columns]
genress.agg(['min','max','mean'])
```

```python
# Adding both the Syntax
genres.agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

Using our logic in apply function but not Built-In 

```python
# Find the number of movies starting with A for each group
def foo(group):
    return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(foo)
```

```python
# Find ranking of each movie in the group according to IMDB Score
def rank_movie(group):
    group['genre_rank'] = group['IMDB_Rating'].rank(ascending = False)
    return group
```

```python
genres.apply(rank_movie)
```

```python
# Find normalised IMDB Rating group (formula is : (x-xmin)/(xmax-xmin))
# def normal(group):
#     group['norm_rating'] = group['IMDB_Rating'] - group['IMDB_Rating'].min()/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
#     return group
# The Above code is showing an error so use new one 
def normal(group):
    min_rating = group['IMDB_Rating'].min()
    max_rating = group['IMDB_Rating'].max()
    if max_rating == min_rating:
        group['norm_rating'] = 0  # or some constant, since normalization doesn't make sense
    else:
        group['norm_rating'] = (group['IMDB_Rating'] - min_rating) / (max_rating - min_rating)
    return group


genres.apply(normal)
```

```python
# find the top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False).head(10).reset_index()
# At last there is no need of reset_index() but I used it to just look output nice but by doing so datatype is changes into dataframe from series
```

```python
# find the batsman with maximum no of sixes
six = ipl[ipl['batsman_runs'] == 6]
six.groupby('batsman')['batsman'].count().sort_values(ascending = False).head(1)
```

```python
# find batsman wiith most number of 4's and 6's in last 5 overs
temp_df = ipl[ipl['over'] > 15]
temp_df = temp_df[(temp_df['batsman_runs'] == 4) | (temp_df['batsman_runs'] == 6) ]
temp_df.groupby('batsman')['batsman'].count().sort_values(ascending = False).head(1)
```

```python
# find Virat Kohli's record against all teams
temp_df = ipl[ipl['batsman'] == 'V Kohli']
temp_df.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# Create a function that can return the highest score of any batsman
temp_df = ipl[ipl['batsman'] == 'V Kohli']
temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending = False).head(1).values[0]
```

```python
def highest(batsman):
    temp_df = ipl[ipl['batsman'] == batsman]
    return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending = False).head(1).values[0]
```
